```python
filepath = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-Coursera/laptop_pricing_dataset_mod1.csv"
df = pd.read_csv(filepath)
```

```python
print(df.info())
```

```python
df[['Screen_Size_cm']] = np.round(df[['Screen_Size_cm']],2)
df.head()
```

```python
# Write your code below and press Shift+Enter to execute 
missing = df.isnull()
for column in df.columns.values.tolist():
    print(column)
    print(missing[column].value_counts())
    print("")
```

```python
# Write your code below and press Shift+Enter to execute
mean = df['Weight_kg'].astype('float').mean(axis=0)
df['Weight_kg'].replace(np.nan, mean, inplace=True)
df['Weight_kg'].head(20)
```

```python
# Write your code below and press Shift+Enter to execute
screen_freq = df['Screen_Size_cm'].value_counts().idxmax()
screen_freq
```

```python
# Write your code below and press Shift+Enter to execute
df[['Screen_Size_cm', 'Weight_kg']].astype('float')
df.dtypes
```

```python
# Write your code below and press Shift+Enter to execute
df['Screen_Size_cm'] = df['Screen_Size_cm'] * 2.54
df['Weight_kg'] = df['Weight_kg'] * 2.205
df.rename(columns={'Screen_Size_cm': 'Screen_Size_inch', 'Weight_kg': 'Weight_lbs'})
```

```python
# Write your code below and press Shift+Enter to execute
df['CPU_frequency'] = df['CPU_frequency']/df['CPU_frequency'].max()
df['CPU_frequency']
```

```python
# Write your code below and press Shift+Enter to execute
bins = np.linspace(df['Price'].min(), df['Price'].max(), 4)
group_names = ['Low', 'Medium', 'High']
df['Price_Binned'] = pd.cut(df['Price'], bins, labels=group_names, include_lowest=True)
df
```

```python
# Write your code below and press Shift+Enter to execute
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot

plt.pyplot.bar(group_names, df["Price_Binned"].value_counts())
plt.pyplot.xlabel("Price")
plt.pyplot.ylabel("count")
plt.pyplot.title("Price bins")
```

```python
y = df['Price']
plt.pyplot.hist(y)
plt.pyplot.ylabel('Price')
plt.pyplot.title('Prices')
```

```python
bins = np.linspace(min(df['Price']), max(df['Price']), 4)
group_names = ['Low', 'Medium', 'High']
df['Price_Pinned'] = pd.cut(df['Price'], bins, labels=group_names, include_lowest=True) 
df.head()
plt.pyplot.hist(df['Price_Pinned'])
```

```python
x = df['Screen_Size_cm']
y = df['Price']
plt.pyplot.xlabel('Screen Size')
plt.pyplot.ylabel('Price')
plt.pyplot.title('Screen Size vs Price')
plt.pyplot.scatter(x, y)
```

```python
dummy_variable = pd.get_dummies(df['Screen'])
dummy_variable.rename(columns= {'Full HD': "Screen-Full_HD", 'IPS Panel': "Screen-IPS_panel"}, inplace=True)
df = pd.concat([df, dummy_variable], axis=1)
df.drop('Screen', axis=1, inplace=True)
```
